I am to demonstrate the motion of a pendulum, shown above, under these given initial conditions:

$$m = 0.15\;\mathrm{Kg}$$
$$c \approx 0.0012\;Ns^2/m^2 $$
$$L = 1.0\;\mathrm{m}$$
$$g = 9.8\;\mathrm{m/s}^2$$

Then add a driving force under the following conditions:

$$ F_D(t) = 0.20\cos(\Omega \ t) $$
For
$$ \Omega = {1.5, 2, 3, 3.13, 3.5, 4, 4.5} $$

I will be using the Runge-Kutta method in the fourth order in order to obtain my data, the associated equations are as follows:

\begin{align}
 k_{1v} & = a_i(x_i, v_i, t_i) \Delta t\\ 
 k_{1x} & = v_i\Delta t\\ 
 k_{2v} & = a(x_i + \frac{k_{1x}}{2}, v_i + \frac{k_{1v}}{2}, t_i + \frac{\Delta t}{2}) \Delta t \\
 k_{2x} & = ( v_i + \frac{k_{1v}}{2}) \Delta t \\
 k_{3v} & = a_i(x_i + \frac{k_{2x}}{2}, v_i + \frac{k_{2v}}{2}, t_i + \frac{\Delta t}{2}) \Delta t \\
 k_{3x} & = ( v_i + \frac{k_{2v}}{2}) \Delta t \\
 k_{4v} & = a(x_i + k_{3x}, v_i + k_{3v}, t_i + \Delta t) \Delta t \\
 k_{4x} & = ( v_i + k_{3v}) \Delta t \\
\end{align}

And I will update the data with these functions, in agreement with the Runge-Kutta method:

\begin{align}
v_{i+1} & = v_i + \frac{1}{6}(k_{1v} + 2k_{2v} + 3k_{3v} + 4k_{4v})\\
x_{i+1} & = x_i + \frac{1}{6}(k_{1x} + 2k_{2x} + 3k_{3x} + 4k_{4x})\\
\end{align}

I will define positive y to be upward and the position of the pendulum at rest to be the origin of my coordinate system. I will be tracking the angle theta in the range $ ( - \pi, \pi). $
To obtain an acceleration function in the case of quadratic damping as required by the Runge-Kutta method, I started with Newton's law, adding a force which resists acceleration in proportion to the square of the velocity. The mathematics works as follows:

$$ \sum F_x = m a_x $$

$$ -mg\sin \theta - c (L\dot{\theta})^2 \hat{\dot{\theta}} = m L\ddot{\theta} $$

Where I have used the fact that angular acceleration is the product of radius of curvature and the radial acceleration for cases without slippage. In the next step I need to divide out the mass. The equation of motion for the quadratically damped pendulum, derived directly from Newton's laws, is now used to obtain the acceration function in the form

$$ a(\theta, \omega, t) = \frac{d\omega}{dt} = - \left(\frac{g}{L}\sin\theta_i + \frac{c L}{m} \omega_i^2 \hat{\omega_i} \right). $$

To add a driving force, simply take the force and divide by the mass in order to obtain the acceleration and then input it into the equation.

$$ a(\theta, \omega, t) = \frac{d\omega}{dt} = - \left(\frac{g}{L}\sin\theta_i + \frac{c L}{m} \omega_i^2 \hat{\omega_i} - \frac{F_D}{m \ L}\right). $$

In the case of linear damping, the math to obtain an expression for acceleration works out similarly, except we make the drag coefficient linearly related to the velocity yeilding

$$ a(\theta, \omega, t) = \frac{d\omega}{dt} = - \left(\frac{g}{L}\sin\theta_i + \frac{c}{m} \omega_i - \frac{F_D}{m \ L}\right). $$

In order to obtain a chaotic motion, I will change my initial conditions to represent a very low damping coefficient, then I will modify the mass of the pendulum. The natural resonant frequency of the pendulum is mass independent so I should see some interesting behavior changes in playing with the mass because the other two terms in my acceleration function are affected by the mass of the pendulum. I will also choose a higher initial angle and then a second initial large angle under the same conditions otherwise to show that my conditions truly do produce chaotic motion. I will not be adjusting the acceleration due to gravity or the length of the pendulum in order to maintain the same resonant frequency throughout this report. I will plot my results throughout my code below.

```python
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import display, Math, Latex

%matplotlib inline
```

```python
"""
The following defines the initial conditions of the pendulum
"""
g = 9.8 # Acceleration due to gravity
m = 0.15 # Mass of the pendulum
L = 1.0 # Length of the pendulum
d = [1.5,2,3,3.13,3.5,4,4.5] # driving freq
c = [0.0012,0.01,1.0,10.0] # damping coefficient

dt = 0.01 # time step
t = np.arange(0,600,dt) # this array represents times, I'm using 600 seconds, 10 minutes, as my maximum

theta_0 = 0.1 # initial angle in radians
omega_0 = np.sqrt(g/L) #initial angular velocity, natural resonant angular freq. 
```

```python
"""
Here I define functions to help fill my data arrays using the Runge-Kutta method
I will need to define one set for non-driven cases and another for driven cases.
"""

def accel(theta, omega, c, m, g, L): 
    return -1*((g/L)**2*np.sin(theta) + (c*L/m) * omega**2 *np.sign(omega))

def drivenAccel(theta, omega, c, m, g, L, d, t):
    return -1*(g*np.sin(theta)/L + (c*L/m) * omega**2 *np.sign(omega) - (0.20*np.cos(d*t))/m*L)

def linAccel(theta, omega, q, m, g, l, d, fD, t):
    return -1*((g/l)*np.sin(theta) + (q)*omega**2/(np.sqrt(g/l))*np.sign(omega) - (fD*np.sin((d*t))))

def rungeKutta4(theta, omega, dt, c, m, g, L):

    k1v = accel(theta, omega, c, m, g, L )*dt
    k1x = omega*dt
    k2v = accel(theta + k1x/2, omega + k1v/2, c, m, g, L )*dt
    k2x = (omega + k1v/2)*dt
    k3v = accel(theta + k2x/2, omega + k2v/2, c, m, g, L)*dt
    k3x = (omega + k2v/2)*dt
    k4v = accel(theta + k3x, omega + k3v, c, m, g, L)*dt
    k4x = (omega + k3v)*dt
    omega = omega + (k1v + 2*k2v + 2*k3v + k4v)/6
    theta = theta + (k1x + 2*k2x + 2*k3x + k4x)/6
   
    return theta, omega

def drivenRungeKutta4(theta, omega, dt, c, m, g, L, d, t):

    k1v = drivenAccel(theta, omega, c, m, g, L, d, t)*dt
    k1x = omega*dt
    k2v = drivenAccel(theta + k1x/2, omega + k1v/2, c, m, g, L, d, t)*dt
    k2x = (omega + k1v/2)*dt
    k3v = drivenAccel(theta + k2x/2, omega + k2v/2, c, m, g, L, d, t)*dt
    k3x = (omega + k2v/2)*dt
    k4v = drivenAccel(theta + k3x, omega + k3v, c, m, g, L, d, t)*dt
    k4x = (omega + k3v)*dt
    omega = omega + (k1v + 2*k2v + 2*k3v + k4v)/6
    theta = theta + (k1x + 2*k2x + 2*k3x + k4x)/6
   
    return theta, omega

def linRungeKutta4(theta, omega, dt, q, m, g, l, d, fD, t):

    k1v = linAccel(theta, omega, q, m, g, l, d, fD, t)*dt
    k1x = omega*dt
    k2v = linAccel(theta + k1x/2, omega + k1v/2, q, m, g, l, d, fD, t)*dt
    k2x = (omega + k1v/2)*dt
    k3v = linAccel(theta + k2x/2, omega + k2v/2, q, m, g, l, d, fD, t)*dt
    k3x = (omega + k2v/2)*dt
    k4v = linAccel(theta + k3x, omega + k3v, q, m, g, l, d, fD, t)*dt
    k4x = (omega + k3v)*dt
    omega = omega + (k1v + 2*k2v + 2*k3v + k4v)/6
    theta = theta + (k1x + 2*k2x + 2*k3x + k4x)/6
   
    return theta, omega

def energy(theta, omega, m, g, L):
    pE = m*g*L*(1 - np.cos(theta))
    kE = (m/2)*(L*omega)**2
    totalEnergy = pE + kE
    
    return pE, kE, totalEnergy
```

```python
"""
Here I will define the appropriate sized data arrays for each value I want to calculate or record
Then I write a loop to fill these arrays with the desired data using my previously defined functions
"""
angle = np.zeros(len(t))
angVel = np.zeros(len(t))
pE = np.zeros(len(t))
kE = np.zeros(len(t))

angle[0] = theta_0
angVel[0] = omega_0
totalEnergy = np.zeros(len(t))
pE[0], kE[0], totalEnergy[0] = energy(angle[0], angVel[0], m, g, L)

i = 0
for i in range(1,len(t)):  
    angle[i], angVel[i]= rungeKutta4(angle[i-1], angVel[i-1], dt, c[0], m, g, L)
    pE[i], kE[i], totalEnergy[i] = energy(angle[i-1], angVel[i-1], m, g, L)
    if angle[i]>np.pi:
        angle[i] = angle[i]-(2*np.pi)
    if angle[i]<(-1*np.pi):
        angle[i] = angle[i]+(2*np.pi)
    i = i+1
```

```python
"""
Here I define another int value in order to clarify iterations between two sets of data
My first set being my arrays of data representing the motion of the pendulum and the second 
representing the maximum angle of each period and times at which the max occurs
"""

maxAngleTimes = []
maxAngle = []
i = 0
for i in range(1,len(t)-1):
    if angle[i] > 0.0:
        if angle[i-1] < angle[i] and angle[i] > angle[i+1]:
            maxAngle.append(angle[i])
            maxAngleTimes.append(t[i])
```

```python
plt.figure(1)
plt.figure(figsize=(15,15))
plt.subplot(311)
plt.plot(t, angle, color = 'red', linewidth= 1.0)
plt.title('The Runge-Kutta Method, 4th order: Angle Versus Time', fontsize = 18)
plt.ylabel('Angle (radians)', fontsize = 16)
plt.grid(True)
plt.xlabel('Time (s)', fontsize = 16)
plt.show()

plt.figure(2)
plt.figure(figsize=(15,15))
plt.subplot(312)
plt.plot(t, angVel, color = 'black', linewidth= 1.0)
plt.title('The Runge-Kutta Method, 4th order: Angular Velocity Versus Time', fontsize = 18)
plt.ylabel('Angular Velocity (m/s)', fontsize = 16)
plt.grid(True)
plt.xlabel('Time (s)', fontsize = 16)
plt.show()

plt.figure(3)
plt.figure(figsize=(15,15))
plt.subplot(313)
plt.plot(angle, angVel , color = 'green', linewidth= 1.0)
plt.title('The Runge-Kutta Method, 4th order: Angular Velocity Versus Angle', fontsize = 18)
plt.ylabel('Angular Velocity (m/s)', fontsize = 16)
plt.xlabel('Angle (radians)', fontsize = 16)
plt.grid(True)
plt.show ()

plt.figure(4)
plt.figure(figsize=(15,5))
kinetic, = plt.plot(t, kE, color='navy', linewidth=1.0)
potential, = plt.plot(t, pE, color='red', linewidth=1.0)
mechanical, = plt.plot(t, totalEnergy, color='black', linewidth=1.0)
plt.ylabel('Energy (J)', size = '20')
plt.title('Energy Versus Time', size = '20')
plt.xlabel('Time (s)', size = '20')
plt.legend([kinetic, potential, mechanical], ["Kinetic Energy", "Potential Energy", "Mechanical Energy"])
plt.grid(True)
plt.show()
```

These plots show my data for the case of an undriven quadratically damped pendulum, the results are certainly in line with expectation as we see both the energy of the system and the amplitude of the oscillations decrease exponentially with time.
